```python
img = cv2.imread('img/adhar_card.jpeg', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((8,8), np.uint8)
_, tR = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY_INV)
dilation = cv2.dilate(tR, kernel)
contours, hier = cv2.findContours(dilation, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img, contours, 11, (0, 255, 0), 1)
cv2.imshow("image", dilation)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
points = cv2.findNonZero(dilation)
rect = cv2.minAreaRect(points)
```

```python
config = ("-l eng --oem 1 --psm 7")
text = pytesseract.image_to_string(img, config=config)
text
```

```python
ekernel = np.ones((2,2), np.uint8)
erosion = cv2.erode(dilation, ekernel, iterations=1)
```
